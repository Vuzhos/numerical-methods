# Теоретическая справка

#### Решаемая проблема

Решение данных уравнений подразумевает два этапа:
1. Локализация корней — выделение отрезков, на которых находится не более одного (возможно кратного) корня;
2. Поиск корня с заданной точностью.

#### Метод Дихотомии
Метод сокращает длину отрезка вдвое с каждой итерацией, откуда следует, что на n-ой итерации длина отрезка будет $(b - a)/2^n$, где a и b - начальная и конечная координаты изначального отрезка соотвественно. Если задана точность $\varepsilon = |x_b - x_*|$, с которой нужно определить корень, то справедливо будет, что $\varepsilon \leq (b - a)/2^n$. Тогда число итерацией для точности $\varepsilon$ оценивается так:
$n = [\log_2{\frac{b - a}{\varepsilon}}] + 1$, где квадратные скобки обозначают целую часть числа

#### Метод простых итераций

Заменим уравнение $f(x) = 0$ эквивалентным ему уравнением $x = \varphi(x)$.
Это можно сделать многими способами, например, положив $\varphi(x) = x + \psi(x)f(x)$, где $\psi(x)$ — произвольная непрерывная знакопостоянная функция. Выберем некоторое нулевое приближение x0 и вычислим дальнейшие приближения по формулам $x_{n+1} = \varphi(x_n),\, n = 0,\, 1,\, 2,\, ...$
Очевидно, что если $\lim\limits_{n \to \infty}{x_n} = x_*$, то $x_*$ - корень исходного уравнения.

Теорема:
Если $\varphi(x)$ липшиц-непрерывна с постоянной $q \in (0,\, 1)$ на отрезке $U_r(a)$, причем $|\varphi(a) - a \leq (1 - q)r|$, то уравнение $x = \varphi(x)$ при любом начальном приближении $x_0 \in U_r(a)$:
1) Имеет на отрезке $U_r(a)$ единственное решение;
2) Метод простой итерации $x_{n + 1} = \varphi(x_n)$ сходится к $x_*$;
3) Для погрешности справедлива следующая оценка:
$|x_k - x_*| \leq q^k |x_0 - x_*|,\ \ \ \ k = 0,\, 1,\, 2,\, ...$

У этой теоремы есть ряд следствий:
1. Если $|\varphi'(x)| \leq q < 1$ для всех $x \in U_r(a)$, выполнено условие $|\varphi(a)| - a \leq (1-q)r$ и $x_0 \in U_r(a)$, то уравнение $x = \varphi(x)$ имеет единственное решение $x_* \in U_r(a)$, метод $x_{n + 1} = \varphi(x_n)$ сходится и справедлива оценка $|x_k - x_*| \leq q^k |x_0 - x_*|,\, k = 0,\, 1,\, 2,\, ...$
2. Пусть уравнение $\varphi(x) = x$ имеет решение $x_*$, функция $\varphi(x)$ непрерывно дифференцируема на отрезке $U_r(x_*) = \{x:\, |x - x_*| \leq r\}$ и $|\varphi'(x_*)| < 1$. Тогда существуют $\varepsilon > 0$ такое, что на отрезке $U_r(x_*)$ уравнение $\varphi(x) = x$ не имеет других решений и метод $x_{n + 1} = \varphi(x_n)$ сходится, если только $x_0 \in U_\varepsilon(x_*)$

Критерий остановки:
$|q\frac{x_n - x_{n - 1}}{1 - q}| = |\frac{(x_n - x_{n - 1})^2}{|2x_{n - 1} - x_n - x_{n - 2}|}| < \varepsilon$

#### Метод Ньютона (метод касательных)
Пусть задано уравнение $f(x) = 0$. Запишем начальную сумму ряда Тейлора: $f(x) \approx f(x_0) + (x - x_0)f'(x_0)$. Заменим в исходном уравнении функцию $f(x)$ полученным приближением:
$f(x_0) + (x - x_0)f'(x_0) = 0$

Если теперь выразить x и сделать ряд замен ($x_0 \to x_n,\, x \to x_{n + 1}$), то получим итерационный процесс:
$x_{n + 1} = x_n - \frac{f(x_n)}{f'(x_n)}$

С помощью второго приближения Тейлора моно получить следующую оценку:
$|x_{n + 1} - x_*| \leq \frac{1}{2}\frac{M_2}{m_1}|x_n - x_*|^2$, где $M_2 = \max\limits_{[a,\, b]} |f''(x)|,\, m_1 = \min\limits_{[a,\, b]} |f'(x)|$

Очевидно, что ошибка на каждом шаге убывает, если:
$\frac{1}{2}\frac{M_2}{m_1}|x_0 - x_*| < 1$, т.е. требуется хорошее начальное приближение.

Теорема.
Пусть задана функция $f(x)$ и определен итерационный процесс $x_{n + 1} = x_n - \frac{f(x_n)}{f'(x_n)}$. Если для всех $x \in [a,\, b]$ справедливо одно из следующих:
1. $f'(x) > 0,\ f''(x) > 0,\ x_0 = b$
2. $f'(x) < 0,\ f''(x) < 0,\ x_0 = b$
3. $f'(x) > 0,\ f''(x) < 0,\ x_0 = a$
4. $f'(x) < 0,\ f''(x) > 0,\ x_0 = a$

### Задание 2
Напишите программу, реализующую нахождение корня одинарной кратности методом деления отрезка пополам

```python
# dichotomy method
def dichotomy(func, interval, eps = 0.001):
    (left, right) = interval 
    if (func(left) * func(right) > 0):
        print("The same signs on ends of a interval!")
        return None
  
    iterations = 0
    mid = 0
    while (abs((left - right) / 2) > eps):
        mid = (left + right) / 2
        iterations += 1
        if (func(left) * func(mid) <= 0):
            right = mid
        else:
            left = mid

    return (mid, iterations)
```

```python
y = lambda x: x**5 - 12 * x**4 + 9 * x**3 + 270 * x**2 - 324 * x - 1944
intervals = [(-5, -2.5), (5, 7.5)]
for interval in intervals:
    print("Interval:", interval)
    root = dichotomy(y, interval, eps = 10**(-4))
    if root != None:
        print("Root:", root[0])
        print("Iterations:", root[1])
    print()
```

Был найден только один корень, поскольку на первом интервале знаки значений функции на концах отрезка одинаковые

### Задание 3-4
Напишите программу нахождения решений уравнения $f(x) = 0$ методом Ньютона и используйте её для поиска всех корней полинома. Для кратных корней использовать модифицированный метод Ньютона. Выведите на экран число итераций.

```python
# function for calculating derivative
def derivative(func, x0, eps = 0.001):
    return (func(x0 + eps) - func(x0 - eps)) / (2 * eps)
```

```python
# Newton's method
def Newton(func, x0, eps = 0.001, max_iterations = 100):
    x_n = x0
    iterations = 0

    while iterations < max_iterations:
        iterations += 1
        f_x = func(x_n)
        
        if not f_x:
            return (x_n, iterations)
        
        x_n_plus_1 = x_n - f_x / derivative(func, x_n, eps)
        
        if abs(x_n_plus_1 - x_n) <= eps:
            break    
        x_n = x_n_plus_1
        
    return (x_n_plus_1, iterations)

# modified Newton's method
def modified_Newton(func, x_0, p, eps = 0.001, max_iterations = 100):
    x_n = x_0
    iterations = 0
    
    while iterations < max_iterations:
        iterations += 1
        f_x = func(x_n)
        
        if not f_x:
            return (x_n, iterations)
        
        x_n_plus_1 = x_n - p * f_x / derivative(func, x_n, eps)
        if abs(x_n_plus_1 - x_n) <= eps:
            break    
        x_n = x_n_plus_1
        
    return (x_n_plus_1, iterations)
```

```python
y = lambda x: x**5 - 12 * x**4 + 9 * x**3 + 270 * x**2 - 324 * x - 1944
eps = 10**(-4)
# roots are close to -2.5 and 5, so let's assume it's a first approximation
assumed_roots = [-2.5, 5]
# we know that multiplicity of -3 is 2 and of 6 is 3
multiplicity = [2, 3]
for i in range(len(assumed_roots)):
    print("First approximation:", assumed_roots[i])
    root_1 = Newton(y, assumed_roots[i], eps)
    print("Root:", root_1[0])
    print("Iterations by simple Newton's method:", root_1[1])
    root_2 = modified_Newton(y, assumed_roots[i], multiplicity[i], eps)
    print("Root:", root_2[0])
    print("Iterations by modified Newton's method:", root_2[1])
    print()
```

### Задание 5
Найдем методом простых итераций корни уравнения $x^2 - a = 0$ (квадратный корень из a).
Приведем уравнение к виду, удобному для использования метода: $x = \frac{1}{2}(\frac{a}{x} + x)$.

```python
# fixed point iteration method
def fixed_point_iteration(func, interval, eps = 0.001, max_iteration = 100):
    x_n = (interval[0] + interval[1]) / 2
    iterations = 0
    
    while iterations < max_iteration:
        x_n_plus_1 = func(x_n)
        iterations += 1
        if abs(x_n - x_n_plus_1) <= eps:
            break
        x_n = x_n_plus_1

    return (x_n_plus_1, iterations)
```

```python
a = 23
print("Our number:", a)
root = fixed_point_iteration(lambda x_n: 1 / 2 * (a / x_n + x_n), (1, 1), 0)
print("Square root by fixed-point iteration:", root[0])
print("Iterations:", root[1])
print("Square root by numpy:", np.sqrt(a))
```

#### Дополнительное задание
Существует множество подвидов метода Ньютона. Например, на лекциях был рассмотрен метод секущих, который позволяется сэкономить время на вычилсениях производной, но из-за чего теряет в скорости сходимости. В этом методе производная вычисляется итеративно по следующей формуле: $f'(x_n) \approx \frac{f(x_n) - f(x_{n - 1})}{x_n - x_{n - 1}}$. 

То есть основная формула имеет вид:
$x_{n + 1} = x_n - f(x_n) \cdot \frac{x_n - x_{n - 1}}{f(x_n) - f(x_{n - 1})}$, но для начала нужно определить два различных приближения $x_0$ и $x_1$.

Но еще больше меня заинтересовал метод Галлея, который позволяет достичь 3 порядка точности, в отличии от классического метода Ньютона. Формула Галлея была получена при рассмотрении одной из вариаций метода Ньютона, которая выглядела так:

$x_{n + 1} = x_n - \frac{f(x_n)}{f'(x_n) - pf(x_n)}$, где $p \in \mathbb{R}$

Исследовав этот метод для разных p было замечено, что в общем случае эта формула дает второй порядок точности. Однако при $p = \frac{f''(x_n)}{2f'(x_n)}$ метод имеет точность третьего порядка. Подставив это p в формулу выше и преобразовав немного получившееся выражение мы получаем метод Галлея:

$x_{n + 1} = x_n - \frac{2f'(x_n)f(x_n)}{2(f'(x_n))^2 - f''(x_n)f(x_n)}$

Но в этом методе есть крайне неприятная операция взятия второй производной, которая может существенно замедлять алгоритм, поэтому я решил поискать оптимизации этого процесса. Один из способов оптимизации заключается в использовании схемы "предиктор-корректор". 

В качестве предиктора (грубого приближения) выступает классический метод Ньютона:

$y_n = x_n - \frac{f(x_n)}{f'(x_n)}$

В качестве корректора (уточнающего шага) выступает метод Галлея:

$x_{n + 1} = y_n - \frac{2f'(y_n)f(y_n)}{2(f'(y_n))^2 - f''(y_n)f(y_n)}$

Теперь мы можем воспользоваться следующей аппроксимацией, которая избавит нас от вычисления второй производной:

$f''(y_n) \approx \frac{f'(y_n) - f'(x_n)}{y_n - x_n}$

Немного расчетов:

$x_{n + 1} = y_n - \frac{2f'(y_n)f(y_n)}{2(f'(y_n))^2 - \frac{f'(y_n) - f'(x_n)}{y_n - x_n}f(y_n)} = y_n - \frac{2f'(y_n)f(y_n)}{2(f'(y_n))^2 + \frac{f'(y_n) - f'(x_n)}{\frac{f(x_n)}{f'(x_n)}}f(y_n)} = y_n - \frac{2f'(y_n)f(y_n)f(x_n)}{2f(x_n)(f'(y_n))^2 + (f'(y_n) - f'(x_n))f'(x_n)f(y_n)} = y_n - \frac{2f'(y_n)f(y_n)f(x_n)}{2f(x_n)(f'(y_n))^2 + f'(x_n)f'(y_n)f(y_n) - (f'(x_n))^2f(y_n)}$

В итоге мы получаем модифицированный двухшаговый алгоритм Галлея, в котором не нужно вычислять вторую производную (порядок точности 3):

$
y_n = x_n - \frac{f(x_n)}{f'(x_n)} \\
x_{n + 1} = y_n - \frac{2f'(y_n)f(y_n)f(x_n)}{2f(x_n)(f'(y_n))^2 + f'(x_n)f'(y_n)f(y_n) - (f'(x_n))^2f(y_n)}
$

Теперь реализуем их программно и сравним с методом Ньютона на примере следующей функции:

$f(x) = x^3 - x^2 - 8x + 12$ 

Проиллюстрируем ее:

```python
x = np.linspace(-5, 5, 100)
y = lambda x: x**3 - x**2 - 8 * x + 12

plt.figure(figsize=(10,5))
plt.plot(x, y(x), c='r')
plt.grid(True)

plt.show()
```

Нетрудно убедиться, что корни будут -3 и 2, причем 2 имеет четную кратность (кратность 2). Проверим наши методы на интервалах (-4, -2) и (1, 3). В качестве первого приближения будем использовать левый конец интервала (для метода секущих в качестве второго приближения буду использовать значение первого + 0.1). Точность будет $\varepsilon = 10^{-6}$

```python
# secant method
def secant(func, x_0, x_1, eps = 0.001, max_iteration = 100):
    x_n_minus_1 = x_0
    x_n = x_1
    iterations = 0

    while iterations < max_iteration:
        x_n_plus_1 = x_n - func(x_n) * (x_n - x_n_minus_1) / (func(x_n) - func(x_n_minus_1))
        iterations += 1
        if abs(x_n_plus_1 - x_n) <= eps:
            break
        x_n_minus_1 = x_n
        x_n = x_n_plus_1
        
    return (x_n_plus_1, iterations)
```

```python
# modified Halley's method
def modified_Halley(func, x_n_plus_1, eps = 0.001, max_iteration = 1000):
    iterations = 0

    while iterations < max_iteration:
        iterations += 1
        
        x_n = x_n_plus_1
        y_n = x_n - func(x_n) / derivative(func, x_n, eps)
        der_x = derivative(func, x_n, eps)
        der_y = derivative(func, y_n, eps)
        f_x = func(x_n)
        f_y = func(y_n)
        if not f_x:
            return (x_n, iterations)
        if not f_y:
            return (y_n, iterations)
        
        x_n_plus_1 = y_n - 2 * f_x * f_y * der_y / (2 * f_x * der_y ** 2 - der_x ** 2 * f_y + der_x * der_y * f_y)
        
        if abs(x_n_plus_1 - x_n) <= eps:
            break
        
    return (x_n_plus_1, iterations)
```

```python
y = lambda x: x**3 - x**2 - 8 * x + 12
eps = 10**(-6)
assumed_roots = [-4, 1]
intervals = [(-4, -2), (1, 3)]
multiplicity = [1, 2]
for i in range(len(assumed_roots)):
    print("First approximation:", assumed_roots[i])
    print("Interval:", intervals[i])
    
    root_1 = Newton(y, assumed_roots[i], eps)
    print("Root by simple Newton method:", root_1[0])
    print("Iterations by simple Newton's method:", root_1[1])
    
    root_2 = modified_Newton(y, assumed_roots[i], multiplicity[i], eps)
    print("Root by modified Newton method:", root_2[0])
    print("Iterations by modified Newton's method:", root_2[1])
    
    root_3 = secant(y, intervals[i][0], intervals[i][0] + 0.1, eps)
    print("Root by secant method:", root_3[0])
    print("Iterations by secant method:", root_3[1])
    
    root_4 = modified_Halley(y, assumed_roots[i], eps)
    if root_4 == None:
        print("This method did not work here!")
    else:
        print("Root by modified Halley's method:", root_4[0])
        print("Iterations by modified Halley's method:", root_4[1])
    print()
```

Делаем вывод, что самым эффективным методом для находения кратных корней (при известной кратности) является модифицированный метод Ньютона, однако в других случаях метод Галлея работает быстрее всех
